- From said colors notes, we could simplify the color definitions as:
    - Primary Colors: The 'default-white' / 'sect-header-bright' notebook theme
    - Secondary Colors: The 'default-black', that complements primary and vice versa

    - Accented Primary/Secondary Colors: The 'alternates'
    - Neutrals: The Shadows

- Issues in our variable naming scheme
    - We tend to name our variables after the elements we're applying them to, which can result in confusion when two elements are 'similarly' colored, but have different palletes. it's better to simplify.
