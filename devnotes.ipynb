Annotations on original styles.css:

General:
- before, after - this is the universal margin-padding

- :root - for background image
- :root - for font size (rem)
- :root - for text color variables + size variables
- idea: use a separate :root for other variables like padding variables etc.

- body - for setting the flex column of the site (for fixing the footer problem)
- h1,h2,h3,h4,h5,h6,p,a,img - for setting 'default text' settings,
    - this might've backfired at times to be honest, especially due to shadow, margin, and color
- h2 / h6 / p - for individual settings
- link-underline /link-underline:hover - for setting underlines of a tags.
    - tracing variablesis sometimes tricky specifically because we have so much
- ul, li, ol - for removing list decorations

Header:
- header - for the navbar container, setup to be sticky
    - note that our header is set up to have text container and the bg container. hmmmge.
- header-container - for the text
- logo h1 / logo h1:hover - for configuring the color of one of the texts in the logo
- logo span - the other color
- nav ul - for flexboxing the nav elements (home, about, projects, blogs)
- nav a / nav a:hover - for configuring nav a tag elements

Footer:
- footer - same as header, for the bg container of footer
- footer-container - for the text
- footer ul - for the footer navbar links (the ext site links)

Main-Aside : <- this is a div container containing both main and aside tags. why? it was to flex them.
- main-aside-container - for the main div

Main:
- main - sets the size of the main side of the main-aside container, to give space for the aside
- section - for grouping and separating sections
- featured-sections - for the featured section tables, flex so that the two containers have space through the gap attribute
- projectsLink, blogsLink - for the a tag container of the projects and blogs in the featured sections.
- featured-section ol - for configuring the ordered list container of the featured section tables
- featured-section li / featured-section li:hover - for configuring list element spacing and hover reaction
- featured-section li:last-child - the last child of the featured section table
- featured-section li h2, featured-section li p / featured section li h2 - setting h2 color settings.
- see-more-link - for setting the see more link to the featured article
- featured-column - for setting the sizes of the featured table columns

- alinks - for _trying_ to set the link alignments
- profile - for setting the profile image params
- social-links - for configuring the list of external website links
