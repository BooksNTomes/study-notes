    - For the rewrite, I suggest we take a tailwind approach and make general layout settings that we can then assign to divs, instead of styling particular divs (main-aside, etc, etc) with exclusive style settings.
        - Example:
            ```css
                .p-1-4-1{
                    padding-top: 1rem;
                    padding-left: 4rem;
                    padding-right: 4rem;
                    padding-bottom: 4rem;
                }
            ```
        - Which is clear imitation of tailwind
    - An alternative approach would be to:
        - Identify Layouts
            - Create Stylings For Them (layout elements) Individually
                - Identify repeated styles
                    - Create repeatably usable style settings
    - The first approach would be an imitation of tailwind that could grow out of scope
    - > While the similar second approach would let us focus on specific sections of the existing layout plan.
